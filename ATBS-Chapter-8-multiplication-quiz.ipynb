```python
# Project: Multiplication quiz without PyInputPlus

import random, time

numberOfQuestions = 10
correctAnswers = 0

for questionNumber in range(numberOfQuestions):
    
    # Pick two random numbers
    num1 = random.randint(0,9)
    num2 = random.randint(0,9)
    
    # Define correct answer
    correct = num1 * num2
    
    # Create prompt from both random numbers
    prompt = '#%s: %s x %s = ' % (questionNumber, num1, num2)
    
    # Validate input and time, give three tries
    for i in range(0,3):
        
        print(prompt)
        
        start = time.time()

        answer = input()

        end = time.time()

        try:
            answer = int(answer)
        except:
            print('Please enter a number')
            continue
        if answer == correct:
            break
        else:
            print('Incorrect, try again.')
            continue
        
        
    # One more input validation as we could still have a non-number answer    
    try:
        answer = int(answer)
    except:
        print('Answer not a number')
        continue
    
    # Did we time out?
    if end - start >= 8:
            print('Out of time!')
    
    # Is the answer NOT equal to the correct answer?
    elif answer != correct:
        print('Incorrect!')
    
    # Otherwise, it must be the correct answer
    else:
        print('Correct!')
        correctAnswers += 1
        
    time.sleep(1) # Brief pause to let user see the result.
    print('Score: %s / %s' % (correctAnswers, numberOfQuestions))
```
